```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crop = pd.read_csv("crop_npk_dataset_final.csv")
```

```python
crop.head()
```

```python
feature_data = crop.drop('label',axis=1)
feature_data.head()
```

```python
corr = feature_data.corr()
corr.head()
```

```python
sns.heatmap(corr,cbar=True,annot=True,cmap="crest")
```

```python
sns.set_theme(style="ticks")
```

```python
sns.pairplot(crop,hue="label")
```

```python
sns.countplot(data=crop,y='label',palette='plasma_r',legend=False)
```

```python
crop.head()
```

```python
sns.boxplot(data=crop,x='label',y='ph')
plt.xticks(rotation=90)
plt.show()
```

```python
crop.head()
```

# Crop vs Rainfall Analysis

```python
sns.set_theme(style="whitegrid")
sns.lineplot(x='label',y='rainfall',data=crop)
plt.title("Crops vs Rainfall")
plt.xlabel("Crops")
plt.ylabel("Rainfall in mm")
plt.xticks(rotation=90)
plt.show()
```

```python
crop.head()
```

# Crop vs pH analysis

```python
sns.lineplot(x='label',y='ph',data=crop)
plt.title("Crop vs pH")
plt.xlabel("Crops")
plt.ylabel("pH")
plt.xticks(rotation=90)
plt.show()
```

# Crop vs Humidity 

```python
sns.lineplot(x='label',y='humidity',data=crop)
plt.title("Crop vs Humidity")
plt.xlabel("Crops")
plt.ylabel("Humidity")
plt.xticks(rotation=90)
plt.show()
```
